```python
import numpy as np
```

```python
bshape = (5, 5)

class Board:
    counter = 0
    
    def __init__(self, numbers, bid = None):
        self.grid = np.array(numbers, dtype=int)
        self.marked = np.zeros_like(self.grid, dtype=bool)
        self.rcounts = np.array([0]*bshape[0])
        self.ccounts = np.array([0]*bshape[0])
        self.haswon = False #not needed, same as bool value of score
        self.score = 0
        self.id = Board.counter
        Board.counter += 1

    def mark(self, n):
        if self.score==0:
            ri, ci = np.where(self.grid==n)
            if len(ri)>0:
                self.marked[ri,ci] = True
                if (bshape[0] in np.sum(self.marked, axis=1)) or (bshape[1] in np.sum(self.marked, axis=0)):
                    self.win(n)
                
    def win(self, n):
        self.score = n * np.sum(self.grid[~self.marked])
        print(f'Board {self.id} won, with score {self.score}.')
```

```python
with open('day4.txt') as f:
    drawn = np.array(f.readline().split(','), dtype=int)

allboards = np.genfromtxt('day4.txt', skip_header=2)
nr = bshape[0]
boards = [Board(allboards[nr*k:nr*(k+1)]) for k in range(len(allboards//5))]
```

```python
for d in drawn:
    for thisboard in boards:
        thisboard.mark(d)
```
