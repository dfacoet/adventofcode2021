```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def chars2bits(s):
    return [int(c) for c in list(s.replace(".", "0").replace("#","1").rstrip())]

with open('day20.txt') as f:
    d = f.readlines()
    replacemap = chars2bits(d[0])
    imagedata = [chars2bits(s) for s in d[2:]]
```

```python
class Image:
    def __init__(self, data, enhancealgo):
        self.image = np.asarray(data, dtype=int)
        self.map = enhancealgo
        self.fill = 0
        self.ensteps = 0
    
    def extend(self):
        zrow = np.full_like(self.image[0],self.fill)
        self.image = np.vstack((zrow, self.image, zrow))
        zcol = np.full((self.image.shape[0],1), self.fill, dtype=int)
        self.image = np.hstack((zcol, self.image, zcol))
    
    def reduce(self):
        #check zeros?
        self.image = self.image[1:-1,1:-1]
    
    bitgrid = 2**np.arange(8,-1,-1).reshape(3,3)
    
    def pixelmap(self, ninegrid):
        return self.map[np.sum(self.bitgrid*ninegrid)]
    
    def enhance(self):
        shape = self.image.shape
        self.extend()
        enhanced = np.zeros_like(self.image)
        self.extend()
        for i in range(shape[0]+2):
            for j in range(shape[1]+2):
                enhanced[i,j] = self.pixelmap(self.image[i:i+3,j:j+3])
        self.image = enhanced
        self.fill = (self.fill + 1) % 2
        self.ensteps += 1
    
    def count(self):
        return self.image.sum()
    
    def plot(self):
        return plt.imshow(self.image, origin='upper')
```

```python
trenchfloor = Image(imagedata, replacemap)
```

```python
trenchfloor.plot()
plt.show()
print(trenchfloor.image.shape, trenchfloor.count())
```

```python
trenchfloor.enhance()
trenchfloor.enhance()
```

```python
trenchfloor.plot()
plt.show()
print(trenchfloor.image.shape, trenchfloor.count())
```

```python
for k in range(48):
    trenchfloor.enhance()
```

```python
trenchfloor.plot()
plt.show()
print(f"After {trenchfloor.ensteps} steps of enhancement, the {trenchfloor.image.shape}-dimensional map has {trenchfloor.count()} lit pixels.")
```
